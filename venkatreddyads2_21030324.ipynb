```python
import wbgapi as wb
```

```python
#Environment indicators
indicator_ids1 = ['EG.ELC.ACCS.ZS','NY.ADJ.SVNG.GN.ZS','AG.LND.FRST.ZS','EN.POP.EL5M.ZS']
country_codes = ['IDN','IND','CHN','JPN','NZL','THA','GBR','IRL','CAN','USA']
my_dataframe1  = wb.data.DataFrame(indicator_ids1, country_codes, mrv=10)
print(my_dataframe1)
#Lofe Expetancy indicators
indicator_ids1 = ['NY.GDP.PCAP.CD','SP.POP.TOTL','SP.RUR.TOTL','SH.DYN.MORT']
country_codes = ['IDN','IND','CHN','JPN','NZL','THA','GBR','IRL','CAN','USA']
my_dataframe2  = wb.data.DataFrame(indicator_ids1, country_codes, mrv=10)
print(my_dataframe2)
```

```python
#Now we are visualizng each of the year values for all indicators of all countries in terms of histogram.
import matplotlib.pyplot as plt
x = my_dataframe1['YR2011']
plt.hist(x)
```

```python
#Applying max values
my_dataframe2.groupby('economy').apply(lambda x: x.astype(float).max())
```

```python
#Replacing or transforming the data based on mean values
my_dataframe1.groupby('economy').transform(lambda x : x.mean())
```

```python
#Pandas groupby function
my_dataframe2.groupby('economy')['YR2014'].mean()
```

```python
#Creation of  a pivot table with pandas
import numpy as np
x= my_dataframe2.pivot_table(index=['series'],aggfunc={'YR2020':np.mean})
y = my_dataframe2.pivot_table(index=['series'],aggfunc={'YR2020':np.median})
print(x,y)
```

```python
#Filtering the data 
my_dataframe1.groupby('economy').filter(lambda x : x['YR2017'].mean() >= 20)
```
